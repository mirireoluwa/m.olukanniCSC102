```python
#project 2

#welcome message
print("Welcome to Mirireoluwa's program")

#importing math
import math

#defining class
class pgmath():
        
    def trapezium(self, height, base1, base2):
        self.height = height
        self.base1 = base1
        self.base2 = base2
        
        #Area of Trapezium Formula
        trap = (self.height/2.0)*(self.base1 + self.base2)
        print(f"Area of Trapezium: {trap}")
        
    def rhombus(self, diagonal1, diagonal2):
        self.diagonal1 = diagonal1
        self.diagonal2 = diagonal2
        
        #Area of Rhombus Formula
        rhom = 0.5 * (self.diagonal1 * self.diagonal2)
        print(f"Area of Rhombus: {rhom}")
        
    def parallelogram(self, base, altitude):
        self.base = base
        self.altitude = altitude
        
        #Area of Parallelogram Formula
        par = base * altitude
        print(f"Area of Parallelogram: {par}")
        
        
    def cube(self, length):
        self.length = length
        
        #Area of Cube Formula
        cubic = 6.0 * (self.length) **(2.0)
        print(f"Area of Cube: {cubic}")
        
        
    def cylinder(self,radius, height1):
        self.height = height1
        self.radius = radius
        
        #Volume of Cylinder Formula
        cyl = math.pi *((self.radius)**2.0) * self.height
        print(f"Volume of Cylinder: {cyl}")
        
#instantiating class
object = pgmath()
name = str(input("Name: "))
formula = str(input("What formula would you like to use?: "))

if formula == "Trapezium":
    a = int(input("Height: "))
    b = int(input("Base 1: "))
    c = int(input("Base 2: "))
    
    object.trapezium(a, b, c)

elif formula == "Rhombus":
    a = int(input("Diagonal 1: "))
    b = int(input("Diagonal 2: "))
    
    object.rhombus(a, b)
    
elif formula == "Parallelogram":
    a = int(input("Base: "))
    b = int(input("Altitude: "))
    
    object.par(a, b)
    
elif formula == "Cube":
    a = int(input("Length: "))
    
    object.cube(a)
    
elif formula == "Cylinder":
    a = int(input("Height: "))
    b = int(input("Radius: "))
    
    object.cylinder(a, b)
    
else:
    print("The formula you have selected is not availalble")
    
print(f"Thank you for using Mirireoluwa's Program {name}")
```
